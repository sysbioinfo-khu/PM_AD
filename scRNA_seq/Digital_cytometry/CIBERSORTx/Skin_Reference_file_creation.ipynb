### Creating reference file for CibersortX from scRNA data

```python
import scanpy as sc
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix
```

```python
# load data
adata = sc.read_h5ad("/scRNA_seq/Digital_cytometry_adata.h5ad")
```

```python
ref_sample = adata.to_df().T
ref_sample
```

```python

ref_sample.columns = adata.obs['Annotation'].values


print(ref_sample.columns.unique())
ref_sample.head()
```

```python
# delete all rows that are filled with 0
ref_sample = ref_sample.loc[(ref_sample!=0).any(axis=1), :]
```

```python
filter_for_genes_in_sample = True

infant_file = "Infant_data.txt"

if filter_for_genes_in_sample:
    
    # read in placenta sample
    bulk = pd.read_csv(infant_file, index_col=0)
    bulk.head()
    print(bulk.columns.values)
    
    print("Filtering for genes in sample")
    ref_sample = ref_sample.loc[ref_sample.index.isin(bulk.columns.values)]

ref_sample
```

```python
print(ref_sample.columns.unique())
```

```python
# rename similar celltypes
ref_sample.columns = ref_sample.columns.str.replace('DC1', 'DC')
ref_sample.columns = ref_sample.columns.str.replace('DC2', 'DC')
ref_sample.columns = ref_sample.columns.str.replace('F1', 'F')
ref_sample.columns = ref_sample.columns.str.replace('F2', 'F')
ref_sample.columns = ref_sample.columns.str.replace('F3', 'F')
ref_sample.columns = ref_sample.columns.str.replace('ILC1_1', 'ILC1')
ref_sample.columns = ref_sample.columns.str.replace('ILC1_3', 'ILC1')
ref_sample.columns = ref_sample.columns.str.replace('ILC1_NK', 'ILC1')
ref_sample.columns = ref_sample.columns.str.replace('LE1', 'LE')
ref_sample.columns = ref_sample.columns.str.replace('LE2', 'LE')
ref_sample.columns = ref_sample.columns.str.replace('VE1', 'VE')
ref_sample.columns = ref_sample.columns.str.replace('VE2', 'VE')
ref_sample.columns = ref_sample.columns.str.replace('VE3', 'VE')
ref_sample.columns = ref_sample.columns.str.replace('Pericyte_1', 'Pericyte')
ref_sample.columns = ref_sample.columns.str.replace('Pericyte_2', 'Pericyte')
```

```python

def sample_columns_by_cell_type(df, x):
    # Get unique column names (cell types)
    unique_cell_types = df.columns.unique()
    
    sampled_indices = []
    
    for cell_type in unique_cell_types:
        # Get indices of columns with this cell type
        cell_type_indices = np.where(df.columns == cell_type)[0]
        cell_type_indices
        
        # Randomly sample x indices (or all if less than x available)
        n_sample = min(x, len(cell_type_indices))
        sampled = np.random.choice(cell_type_indices, size=n_sample, replace=False)
        
        sampled_indices.extend(sampled)
    
    # Sort the indices to maintain the original column order
    sampled_indices.sort()
    
    # Return the dataframe with only the sampled columns
    return df.iloc[:, sampled_indices]

# Use the function
result = sample_columns_by_cell_type(ref_sample, x=400)  # Replace 3 with your desired x
```
